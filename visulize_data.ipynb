```python
import json
import matplotlib as mpl
import matplotlib.pyplot as plt

# Open the JSON file in read mode ("r")
with open("data.json", "r") as file:
    # Load data from the file
    data = json.load(file)
fig = plt.figure(figsize=(14,7))
ax = fig.add_subplot()
# Create a figure and an axis
color_lst = ["red", "blue", "green", "yellow", "black", "orange", "purple", "pink", "brown", "gray"]
for pop_index, popultaion in enumerate(data):
        #print(popultaion)
        i = 0
        x = []
        y = []
        for popultation_values in (data[popultaion]):
            #print(data[popultaion][i])
            mutation_rate = data[popultaion][i].get("mutationrate")
            logged = (data[popultaion][i].get("Logged"))
            i += 1
            x.append(logged[0])
            y.append(logged[1])
            ax.plot(x, y, color=color_lst[pop_index], label=popultaion + " " + str(mutation_rate))
ax.legend()
ax.set_xlabel("Generation")
ax.set_ylabel("Distance")
ax.set_title("Distance vs Generation")
plt.savefig(f"{popultaion}.png".format(popultaion))
plt.show()


```
